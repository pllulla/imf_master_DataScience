# Ejemplo Regresión Lineal Simple

```python
import numpy as np
import random

# Librerías para obtener métricas y la implementación de la regresión lineal
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


import matplotlib.pyplot as plt
```

```python
# Generador de distribución de datos para regresión lineal simple
def generador_datos_simple(beta, muestras, desviacion):
    # Genero n (muestras) valores de x aleatorios entre 0 y 100
    x = np.random.random(muestras) * 100
    # Genero un error aleatorio gaussiano con desviación típica (desviacion)
    e = np.random.randn(muestras) * desviacion
    # Obtengo el y real como x*beta + error
    y = x * beta + e
    return x.reshape((muestras,1)), y.reshape((muestras,1))
```

```python
# Parámetros de la distribución
desviacion = 200
beta = 10
n = 50
x, y = generador_datos_simple(beta, n, desviacion)
```

```python
# Represento los datos generados
plt.scatter(x, y)
plt.show()
```

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

En el siguiente paso se crea un modelo de regresión lineal y se ajusta o entrena utilizando los datos (X,Y) disponibles. De este modo se obtiene una estimación de f(X). Cuando se llama al método predict(x) lo que se hace es aplicar esa función estimada para obtener predicciones o estimaciones de la respuesta Y a partir de las variables independientes X

```python
# Creo un modelo de regresión lineal
modelo = LinearRegression()
```

```python
%%time
# Entreno el modelo con los datos (X,Y)
modelo.fit(x, y)
```

```python
# Ahora puedo obtener el coeficiente b_1
print('Coeficiente beta1: ', modelo.coef_[0])
```

```python
# Podemos predecir usando el modelo
y_pred = modelo.predict(x)
```

```python
# Por último, calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y, y_pred))
print('Estadístico R_2: %.2f' % r2_score(y, y_pred))
```

```python
# Representamos el ajuste (rojo) y la recta Y = beta*x (verde)
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```

```python
#EJERCICIO--------------------------------

import pandas as pd

# Ruta del archivo
file_path = 'tvmarketing.csv'

# Cargar el archivo CSV
data = pd.read_csv(file_path)

# Mostrar las primeras filas del DataFrame para verificar que se ha cargado correctamente
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

X = data['TV'].values.reshape(-1,1)  # La entrada debe ser una matriz 2D, por lo que usamos reshape
y = data['Sales'].values

linear_reg = LinearRegression()

# Entreno el modelo con los datos (X,Y)
linear_reg.fit(X, y)

# Realizar predicciones
y_pred = linear_reg.predict(X)

# Calculamos el error cuadrático medio y el estadístico R^2
print('Error cuadrático medio: %.2f' % mean_squared_error(y, y_pred))
print('Estadístico R_2: %.2f' % r2_score(y, y_pred))

# Representamos el ajuste (rojo) y la recta Y = beta*x (verde)
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
# Valores de TV para los cuales queremos predecir las ventas
tv_values = np.array([[177], [310]])

# Realizar predicciones utilizando el modelo entrenado
predicted_sales = linear_reg.predict(tv_values)

# Resultados de las predicciones
predicted_sales
```
