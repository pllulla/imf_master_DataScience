# Regresión con Vecinos Próximos KNN

```python
# Regresión lineal con KNN
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.metrics   import mean_squared_error, r2_score

import numpy as np

import matplotlib.pyplot as plt

# Generación de datos aleatorios
np.random.seed(0)
X = np.sort(20 * np.random.rand(200,1), axis=0)
y = np.sin(X)[:, 0]
y_ruido = np.sin(X)[:, 0] + np.random.normal(size=X.shape[0],scale=0.20)
```

```python
plt.xlabel("Atributo X")
plt.ylabel("Target")
plt.scatter(X, y_ruido.ravel())
plt.show()
```

### Implementamos un modelo basado en vecinos próximos (KNN) de tipo regresor

Documentación https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html

En este caso, vamos a implementar un modelo basado en 3 vecinos

```python
# Definimos el modelo con 3 vecinos
knn_model = KNeighborsRegressor(n_neighbors = 3)
```

```python
# Lo entrenamos
knn_model.fit(X, y_ruido)
```

```python
# Generamos nuevos datos de test
X_axis = np.linspace(X.min(),X.max(), 200).reshape(200, 1)
```

```python
# Realizamos predicciones sobre el modelo knn
y_axis_pred = knn_model.predict(X_axis)
```

```python
# Mostramos las predicciones
plt.xlabel("Atributo X")
plt.ylabel("Target")
plt.scatter(X, y_ruido.ravel())
plt.plot(X_axis, y_axis_pred, c="red")
plt.show()
```

```python
# Mostramos las métricas
print('Error cuadrático medio: %.2f' % mean_squared_error(y, y_axis_pred))
print('Estadístico R_2: %.2f'        % r2_score(y, y_axis_pred))
```

## ¿Cómo podemos seleccionar los mejores parámetros para evitar el sobreajuste?

Como hemos podido comprobar a medida que el nº de vecinos aumenta, estamos perdiiendo R2 en test y, aumantando demasiado en train, por lo tanto, nuestro modelo está afectado por **Overfitting** o sobreajuste, no obstante, hay muchos parámetros diferentes en KNN (y en todos los modelos) ¿Cómo podemos estar seguros de seleccionar los mejores parámetros del modelo? La respuesta, **validación cruzada** y **selección de hyperparámetros** empleando `GridSearchCV`.

Realizamos el mismo ejercicio, pero ahora *tuneando* hyperparámetros.

```python
# Para realizar validación cruzada y parameter tunning
from sklearn.model_selection import GridSearchCV

# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
```

```python
params = {
    "n_neighbors" : range(2, 10)
}
```

```python
knn_model = KNeighborsRegressor() # Parámetros vacíos
```

```python
# grid_model = GridSearchCV(knn_model, 
#                           params, 
#                           scoring="r2", # https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
#                           cv = 4)


grid_model = GridSearchCV(knn_model, 
                          params, 
                          cv = 3, 
                          scoring= "r2")
```

```python
grid_model.fit(X, y_ruido)
```

```python
grid_model.best_params_
```

```python
# Realizamos predicciones sobre el modelo knn
y_axis_pred = grid_model.best_estimator_.predict(X_axis)
```

```python
# Mostramos las predicciones
plt.xlabel("Atributo X")
plt.ylabel("Target")
plt.scatter(X, y_ruido.ravel())
plt.plot(X_axis, y_axis_pred, c="red")
plt.show()
```

```python
# Mostramos las métricas
print('Error cuadrático medio: %.2f' % mean_squared_error(y, y_axis_pred))
print('Estadístico R_2: %.2f'        % r2_score(y, y_axis_pred))
```
