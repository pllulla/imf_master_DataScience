# Ejercicio sobre algoritmos regresores

## Datos sobre gatos 

🐱 Objetivo: Vamos a modelar la posible relación existente entre parámetros relacionados con la anatomía de los gatos como su peso corporal, sexo y, el peso del corazón del gato.

Las variables disponibles son las siguientes:

1. Bwt: Peso del animal en Kg.
2. Hwt: Peso del corazón en gramos
3. Sex: M/F (male/female)

El objetivo principal es predecir si, dado el sexo de un gato y, el peso corporal del mismo, obtener el peso de su corazón, por lo tanto, **Hwt** es nuestra target.  Para realizar este ejercicio deben aplicarse los siguientes pasos:
* EDA
* Revisar la relación existente entre las variables.
* Pre-Procesamiento
* Aplicar los siguientes algoritmos.
    * Linear Model
    * Support Vector Regressor
    * Random Forest Regressor
    * Decission Tree Regressor
    * K Nearest Neighbors Regressor
    * xGbooks Regressor
* Comparar los modelos en base al R2 y el MSE.

## Solución

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
```

## EDA

Como podemos comprobar, existe una relación lineal entre el peso y el peso del corazón, podemos comprobar también que tanto el peso del corazón y peso corporal en gatos hembras es menor (o no tenemos datos con mayor amplitud). VAmos a analizar la matriz de correlación

```python
sns.heatmap(df.corr(), annot=True)
```

Ambas variables tienen una gran correlación, pero, también tienen diferentes escalas, vamos a normalizarlas.

## Pre-Procesamiento

Solo normalizamos el peso corporal del animal, no hace falta normalizar la target

```python
sc = StandardScaler()
df.Bwt = sc.fit_transform(df.Bwt.values.reshape(-1,1))
```

```python
df.head()
```

Convertimos Sex a números

```python
df.Sex.replace({
    "F" : 1,
    "M" : 0
}, inplace = True)
```

```python
X = df.drop(["Bwt"], 1, inplace = False)
y = df.Bwt
```

## Modelos

**Posibles mejoras**: Emplear GridSearchCV

```python
%%time

svr_model = SVR()
svr_model.fit(X, y)

lm_model = LinearRegression()
lm_model.fit(X, y)

rf_model = RandomForestRegressor()
rf_model.fit(X, y)

knn_model = KNeighborsRegressor()
knn_model.fit(X, y)

tree_model = DecisionTreeRegressor()
tree_model.fit(X, y)

xgb_model = xgb.XGBRegressor()
xgb_model.fit(X, y)
```

## Predictions and model comparison

```python
all_preds = {
    "SVR"                   : svr_model.predict(X),
    "LinearRegression"      : lm_model.predict(X),
    "RandomForestRegressor" : rf_model.predict(X),
    "KNeighborsRegressor"   : knn_model.predict(X),
    "DecisionTreeRegressor" : tree_model.predict(X),
    "XGBRegressor"          : xgb_model.predict(X)
}
```

```python
metrics_r     = []
metrics_error = []

for model in all_preds.keys():
    metrics_r.append(round(r2_score(y, all_preds[model]) ,2))
    metrics_error.append(round(mean_squared_error(y, all_preds[model]) ,2))
```

```python
df_metrics = pd.DataFrame({
    "model_name" : all_preds.keys(),
    "R2 Score"   : metrics_r,
    "MSE"        : metrics_error
})
```

```python
df_metrics
```

```python
plt.bar(df_metrics.model_name, df_metrics.MSE)
for i in range(df_metrics.shape[0]):
    plt.text(df_metrics.model_name[i], 
             df_metrics.MSE[i], str(round(df_metrics.MSE[i],2)))
    
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.bar(df_metrics.model_name, df_metrics["R2 Score"])
for i in range(df_metrics.shape[0]):
    plt.text(df_metrics.model_name[i], 
             df_metrics["R2 Score"][i], str(round(df_metrics["R2 Score"][i],2)))
    
plt.xticks(rotation = 90)
plt.show()
```

En un escenario como este, con parámetros por defecto, podríamos tomar árboles de decisión o XGBoost
