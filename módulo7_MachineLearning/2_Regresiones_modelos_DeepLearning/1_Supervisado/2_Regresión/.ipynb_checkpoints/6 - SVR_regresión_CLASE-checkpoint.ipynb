# Regresión con Máquinas de Soporte Vectorial

```python
# Importo librerías
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
```

```python
# Generación de datos de muestra
X = np.sort(10 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel()
# Añado ruido a la salida
y[::5] += 2 * (0.5 - np.random.rand(20))
```

```python
plt.scatter(X, y, color = 'blue')   
plt.xlabel('Variable X') 
plt.ylabel('Variable Y') 
plt.show()
```

En el siguiente código se generan tres modelos diferentes: uno con un kernel rbf, otro con kernel lineal y otro con kernel polinómico de grado tres. Los datos siguen una forma no lineal, por lo que, previsiblemente el kernel lineal no se comporte correctamente.

https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

```python
# Ajuste del modelo de regresión
svr_rbf = SVR(kernel='rbf',    C=100, gamma=0.1, epsilon=.1)

svr_lin = SVR(kernel='linear', C=100, gamma='auto')

svr_poly = SVR(kernel='poly',  C=100, gamma='auto', degree=3, epsilon=.1,
               coef0=1)
```

```python
# Análisis de los resultados
lw = 2
```

```python
svrs = [svr_rbf, svr_lin, svr_poly]
kernel_label  = ['RBF', 'Linear', 'Polynomial']
model_color   = ['m', 'c', 'g']
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5), sharey=True)

for ix, svr in enumerate(svrs):
    
    axes[ix].plot(X, svr.fit(X, y).predict(X), color=model_color[ix], lw=lw,
                  label='{} model'.format(kernel_label[ix]))
    
    
    axes[ix].scatter(X[svr.support_], y[svr.support_], facecolor="none",
                     edgecolor=model_color[ix], s=50,
                     label='{} support vectors'.format(kernel_label[ix]))
    
    axes[ix].scatter(X[np.setdiff1d(np.arange(len(X)), svr.support_)],
                     y[np.setdiff1d(np.arange(len(X)), svr.support_)],
                     facecolor="none", edgecolor="k", s=50,
                     label='other training data')
    
    axes[ix].legend(loc='upper center', bbox_to_anchor=(0.5, 1.1),
                    ncol=1, fancybox=True, shadow=True)

fig.text(0.5, 0.04, 'Variable X', ha='center', va='center')
fig.text(0.06, 0.5, 'Variable Y', ha='center', va='center', rotation='vertical')
fig.suptitle("Support Vector Regression", fontsize=14)
plt.show()
```

Se cumple lo que se preveía, el kernel lineal tiene un mal rendimiento en estos datos. El kernel polinómico de grado tres en una parte de los datos se ajusta correctamente y en otra no. Por último, el kernel RBF o Gaussiano se ajusta muy bien a los datos en toda la forma. Como norma general se debería utilizar el kernel RBF ya que se ajusta correctamente a casi cualquier función

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# RBF
svr_rbf.fit(X, y)
y_pred_rbf = svr_rbf.predict(X)

print("MSE --> ", round(mean_squared_error(y, y_pred_rbf) ,2))
print("R2 ---> ", round(r2_score(y, y_pred_rbf) , 2))
```

```python
# LINEAL
svr_lin.fit(X, y)
y_pred_lin = svr_lin.predict(X)

print("MSE --> ", round(mean_squared_error(y, y_pred_lin) ,2))
print("R2 ---> ", round(r2_score(y, y_pred_lin) , 2))
```

```python
# POLINOMIAL
svr_poly.fit(X, y)
y_pred_poly = svr_poly.predict(X)

print("MSE --> ", round(mean_squared_error(y, y_pred_poly) ,2))
print("R2 ---> ", round(r2_score(y, y_pred_poly) , 2))
```

Probad, diferentes configuraciones de `GridSearchCV` para ver cuál es el mejor resultado posible (en clase)

```python
from sklearn.model_selection import GridSearchCV
```

```python
# params = {
#     "kernel"  : ["linear", "rbf", "sigmoid"],
#     "gamma"   : (0.001, 0.01, 0.1, 1.0, 2.0),
#     "C"       : np.linspace(0.00001, 6, 60),
#     "epsilon" : np.linspace(0.0001, 1, 15)
# }

params = {
    "kernel"  : ["rbf"],
    "C"       : np.linspace(100, 250, 1),
    "epsilon" : np.linspace(0.000001, 0.0001, 15)
}
```

```python
reg_svr = SVR(gamma = 0.1)
```

```python
grid_svr = GridSearchCV(reg_svr, params, cv = 3)

grid_svr
```

```python
%%time

grid_svr.fit(X, y)
```

```python
grid_svr.best_estimator_
```

```python
grid_svr.best_params_
```

```python
y_grid_svr_pred = grid_svr.best_estimator_.predict(X)
```

```python
print("MSE --> ", round(mean_squared_error(y, y_grid_svr_pred) ,2))
print("R2 ---> ", round(r2_score(y,           y_grid_svr_pred) ,2))
```
