```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n para visualizaci√≥n
sns.set(style="whitegrid")

# Creando una lista de las caracter√≠sticas num√©ricas
numerical_features = [
    "Administrative", "Administrative_Duration", "Informational", 
    "Informational_Duration", "ProductRelated", "ProductRelated_Duration", 
    "BounceRates", "ExitRates", "PageValues", "SpecialDay"
]

# Creando histogramas y boxplots para las caracter√≠sticas num√©ricas
fig, axes = plt.subplots(len(numerical_features), 2, figsize=(15, 4 * len(numerical_features)))

for i, col in enumerate(numerical_features):
    # Histograma
    sns.histplot(data[col], ax=axes[i, 0], kde=True)
    axes[i, 0].set_title(f'Histograma de {col}')
    axes[i, 0].set_xlabel(col)
    axes[i, 0].set_ylabel('Frecuencia')

    # Boxplot
    sns.boxplot(x=data[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot de {col}')
    axes[i, 1].set_xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Revisando valores faltantes y valores √∫nicos en cada columna
missing_values = data.isnull().sum()
unique_values = data.nunique()

# Eliminando columnas con un solo valor √∫nico (si las hay) ya que no aportan informaci√≥n √∫til para el modelo
columns_to_drop = unique_values[unique_values <= 1].index.tolist()
data_cleaned = data.drop(columns=columns_to_drop)

# Justificaci√≥n para la eliminaci√≥n de columnas
justification = f"Se han eliminado las siguientes columnas por tener un √∫nico valor √∫nico, lo cual no aporta informaci√≥n variada para el modelo: {columns_to_drop}" if columns_to_drop else "No se han encontrado columnas con un √∫nico valor √∫nico para eliminar."

justification, missing_values, unique_values
```

```python
# Eliminando filas con valores faltantes usando dropna()
data_cleaned_dropna = data_cleaned.dropna()

# Comprobando si todav√≠a hay valores faltantes despu√©s de la eliminaci√≥n
remaining_missing_values = data_cleaned_dropna.isnull().sum()

# An√°lisis de Outliers
# Para el an√°lisis de outliers, revisaremos las estad√≠sticas descriptivas y los boxplots
outliers_description = data_cleaned_dropna.describe()

# Visualizando los boxplots para identificar outliers
fig, axes = plt.subplots(len(numerical_features), 1, figsize=(10, 4 * len(numerical_features)))
for i, col in enumerate(numerical_features):
    sns.boxplot(x=data_cleaned_dropna[col], ax=axes[i])
    axes[i].set_title(f'Boxplot de {col} despu√©s de eliminar valores faltantes')
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()

remaining_missing_values, outliers_description
```

```python
from sklearn.preprocessing import LabelEncoder

# Copia del dataset para las transformaciones
data_for_modeling = data_cleaned_dropna.copy()

# Lista de caracter√≠sticas categ√≥ricas
categorical_features = ['Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend']

# Aplicando Label Encoding a las caracter√≠sticas categ√≥ricas
# Label Encoding es adecuado cuando las caracter√≠sticas categ√≥ricas tienen un orden natural o cuando se utilizan modelos como redes neuronales
label_encoder = LabelEncoder()
for col in categorical_features:
    data_for_modeling[col] = label_encoder.fit_transform(data_for_modeling[col])

# Verificando las transformaciones
transformed_data_head = data_for_modeling.head()

transformed_data_head
```

```python
from sklearn.preprocessing import StandardScaler

# Separando las caracter√≠sticas y la etiqueta (target)
X = data_for_modeling.drop('Revenue', axis=1)
y = data_for_modeling['Revenue']

# Estandarizaci√≥n de las caracter√≠sticas
# Se utiliza StandardScaler que estandariza las caracter√≠sticas restando la media y dividi√©ndolas por la desviaci√≥n est√°ndar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convertir el resultado en un DataFrame para mejor visualizaci√≥n
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Visualizando las primeras filas del conjunto de datos estandarizado
scaled_data_head = X_scaled_df.head()

scaled_data_head
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# N√∫mero de caracter√≠sticas a seleccionar
# Como punto de partida, elegimos la mitad de las caracter√≠sticas disponibles
k = len(X_scaled_df.columns) // 2

# Utilizando el m√©todo SelectKBest con la prueba ANOVA F-value
selector = SelectKBest(score_func=f_classif, k=k)
X_selected = selector.fit_transform(X_scaled_df, y)

# Obteniendo los √≠ndices de las caracter√≠sticas seleccionadas
selected_features_idx = selector.get_support(indices=True)
selected_features_names = X_scaled_df.columns[selected_features_idx]

# Creando un DataFrame con las caracter√≠sticas seleccionadas para mejor visualizaci√≥n
X_selected_df = pd.DataFrame(X_selected, columns=selected_features_names)

# Visualizaci√≥n de las caracter√≠sticas seleccionadas
selected_features_names, X_selected_df.head()
```

```python
from sklearn.model_selection import train_test_split

# Dividiendo los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y, test_size=0.25, random_state=42)

# Tama√±os de los conjuntos de entrenamiento y prueba para confirmar la divisi√≥n
train_size = X_train.shape, y_train.shape
test_size = X_test.shape, y_test.shape

train_size, test_size
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score
import numpy as np

# Configurando los modelos con sus respectivos hiperpar√°metros para GridSearch
# 1. Regresi√≥n Log√≠stica
logistic_params = {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'solver': ['liblinear', 'lbfgs']}
logistic_model = LogisticRegression()

# 2. Redes Neuronales (MLP Classifier)
mlp_params = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['tanh', 'relu'],
    'alpha': [0.0001, 0.001, 0.01]
}
mlp_model = MLPClassifier(max_iter=1000)

# 3. Random Forest
random_forest_params = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20, 30]}
random_forest_model = RandomForestClassifier()

# Realizando GridSearch con Cross-Validation para cada modelo
# Nota: Se utiliza una validaci√≥n cruzada de 5 pliegues
models = {
    'Logistic Regression': (logistic_model, logistic_params),
    'MLP Classifier': (mlp_model, mlp_params),
    'Random Forest': (random_forest_model, random_forest_params)
}

best_estimators = {}

for model_name, (model, params) in models.items():
    grid_search = GridSearchCV(model, params, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_estimators[model_name] = grid_search.best_estimator_

    # Mostrando los mejores par√°metros para cada modelo
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best cross-validation score (accuracy): {grid_search.best_score_:.4f}")

best_estimators
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluaci√≥n de los modelos en el conjunto de prueba
evaluations = {}

for model_name, model in best_estimators.items():
    # Predicci√≥n en el conjunto de prueba
    y_pred = model.predict(X_test)

    # Calculando la precisi√≥n y generando el informe de clasificaci√≥n y la matriz de confusi√≥n
    accuracy = accuracy_score(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)

    evaluations[model_name] = {
        'Accuracy': accuracy,
        'Classification Report': class_report,
        'Confusion Matrix': conf_matrix
    }

evaluations
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import pandas as pd

# Recargando los datos y preparando nuevamente el conjunto de entrenamiento y prueba
file_path = 'online_shoppers_intention.csv'
data = pd.read_csv(file_path)
data_cleaned = data.dropna()
data_for_modeling = data_cleaned.copy()

# Aplicando Label Encoding a las caracter√≠sticas categ√≥ricas
categorical_features = ['Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend']
label_encoder = LabelEncoder()
for col in categorical_features:
    data_for_modeling[col] = label_encoder.fit_transform(data_for_modeling[col])

# Estandarizaci√≥n de las caracter√≠sticas
X = data_for_modeling.drop('Revenue', axis=1)
y = data_for_modeling['Revenue']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# Selecci√≥n de variables
selector = SelectKBest(score_func=f_classif, k=8)
X_selected = selector.fit_transform(X_scaled_df, y)
selected_features_idx = selector.get_support(indices=True)
selected_features_names = X_scaled_df.columns[selected_features_idx]
X_selected_df = pd.DataFrame(X_selected, columns=selected_features_names)

# Dividiendo los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y, test_size=0.25, random_state=42)

# Diccionario para almacenar los modelos con sus mejores par√°metros
best_estimators = {
    'Logistic Regression': LogisticRegression(C=1, solver='liblinear'),
    'MLP Classifier': MLPClassifier(alpha=0.01, hidden_layer_sizes=(50,), max_iter=1000),
    'Random Forest': RandomForestClassifier(max_depth=10, n_estimators=50)
}

# Evaluando cada modelo en el conjunto de prueba
evaluation_results = {}
for model_name, model in best_estimators.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculando m√©tricas de evaluaci√≥n
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    confusion = confusion_matrix(y_test, y_pred)

    evaluation_results[model_name] = {
        'Accuracy': accuracy,
        'Classification Report': report,
        'Confusion Matrix': confusion
    }

evaluation_results['Logistic Regression']['Accuracy'], evaluation_results['MLP Classifier']['Accuracy'], evaluation_results['Random Forest']['Accuracy']
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Diccionario para almacenar los modelos con sus mejores par√°metros
best_estimators = {
    'Logistic Regression': LogisticRegression(C=1, solver='liblinear'),
    'MLP Classifier': MLPClassifier(alpha=0.01, hidden_layer_sizes=(50,), max_iter=1000),
    'Random Forest': RandomForestClassifier(max_depth=10, n_estimators=50)
}

# Evaluando cada modelo en el conjunto de prueba
evaluation_results = {}
for model_name, model in best_estimators.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculando m√©tricas de evaluaci√≥n
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    confusion = confusion_matrix(y_test, y_pred)

    evaluation_results[model_name] = {
        'Accuracy': accuracy,
        'Classification Report': report,
        'Confusion Matrix': confusion
    }

evaluation_results['Logistic Regression']['Accuracy'], evaluation_results['MLP Classifier']['Accuracy'], evaluation_results['Random Forest']['Accuracy']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score, recall_score, accuracy_score

# Evaluar cada modelo utilizando ROC AUC en validaci√≥n cruzada
cv_results_roc = {}
for model_name, model in best_estimators.items():
    roc_auc = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc').mean()
    cv_results_roc[model_name] = roc_auc

# Entrenar los modelos y evaluarlos en el conjunto de prueba
test_results = {}
for model_name, model in best_estimators.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Calcular m√©tricas en el conjunto de prueba
    accuracy = accuracy_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)

    test_results[model_name] = {
        'Accuracy': accuracy,
        'Recall': recall,
        'ROC AUC': roc_auc
    }

# Mostrar los resultados
print("Resultados de ROC AUC en Cross-Validation:", cv_results_roc)
print("Resultados en el conjunto de prueba:", test_results)


```

```python
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier

# Asumiendo que X_train, y_train, X_test, y_test ya est√°n definidos y son tus conjuntos de datos

# Reconstruir el modelo Random Forest con los par√°metros √≥ptimos
model_rf = RandomForestClassifier(max_depth=10, n_estimators=50)
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)

# Matriz de confusi√≥n
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)

# Extraer TP, TN, FP, FN de la matriz de confusi√≥n
TP = conf_matrix_rf[1, 1]
TN = conf_matrix_rf[0, 0]
FP = conf_matrix_rf[0, 1]
FN = conf_matrix_rf[1, 0]

# Calcular la rentabilidad
rentabilidad = TP * 100 + TN * 100 - FP * 20 - FN * 500

print("Matriz de Confusi√≥n:\n", conf_matrix_rf)
print("Rentabilidad del Modelo: ‚Ç¨", rentabilidad)
```
