```python
# Tratamiento de valores faltantes
data_cleaned = data.dropna()

# Comparación del tamaño del dataset antes y después de eliminar los valores faltantes
original_size = data.shape
cleaned_size = data_cleaned.shape

original_size, cleaned_size
```

```python
# Identificación de outliers utilizando el rango intercuartil (IQR)

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Columnas a verificar para outliers
columns_to_check = ['Administrative_Duration', 'Informational_Duration', 
                    'ProductRelated_Duration', 'PageValues']

# Detección de outliers
outliers_info = {column: detect_outliers(data_cleaned, column).shape[0] for column in columns_to_check}
outliers_info
```

```python
# Aplicación de One-Hot Encoding a las variables categóricas seleccionadas

# Variables para One-Hot Encoding
categorical_vars = ['Month', 'VisitorType', 'OperatingSystems', 'Browser', 'Region']

# Aplicando One-Hot Encoding
data_dummies = pd.get_dummies(data_cleaned, columns=categorical_vars)

# Mostrando las primeras filas del dataset resultante
data_dummies.head()
data_dummies.shape
```

```python
# Análisis de correlación
correlation_matrix = data_dummies.corr()

# Visualización de la matriz de correlación para identificar posibles multicolinealidades
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()

# Revisión de la varianza de las variables
variance = data_dummies.var().sort_values()
low_variance = variance[variance < 0.01]  # Umbral de varianza baja

low_variance
```

```python
# Eliminación de variables con baja varianza y variables originales redundantes

# Variables a eliminar
variables_to_remove = low_variance.index.tolist()

# Eliminación de las variables seleccionadas
data_dummies_reduced = data_dummies.drop(columns=variables_to_remove)

# Mostrando las primeras filas del dataset resultante
data_dummies_reduced.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Seleccionando solo las variables numéricas (excluyendo las dummies)
numeric_columns = data_dummies_reduced.select_dtypes(include=['float64', 'int64']).columns

# Estandarización de las variables numéricas
scaler = StandardScaler()
data_dummies_reduced[numeric_columns] = scaler.fit_transform(data_dummies_reduced[numeric_columns])

# Mostrando las primeras filas del conjunto de datos estandarizado
data_dummies_reduced.head()
```

```python
# Correlación entre la variable objetivo 'Revenue' (convertida a numérica) y las demás variables
target_correlation = data_dummies_reduced.corrwith(data_dummies_reduced['Revenue']).sort_values()

target_correlation
```

```python
from sklearn.model_selection import train_test_split

# Definiendo la variable objetivo y las características
X = data_dummies_reduced.drop('Revenue', axis=1)  # Características
y = data_dummies_reduced['Revenue']               # Variable objetivo

# Dividiendo los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Verificación de las dimensiones de los conjuntos resultantes
(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, KFold

# Configuración del modelo de Regresión Logística
logistic_model = LogisticRegression(max_iter=1000, random_state=42)

# Configuración de K-Fold Cross-Validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Ejecución de la Validación Cruzada
cross_val_scores = cross_val_score(logistic_model, X_train, y_train, cv=kf, scoring='accuracy')

# Calculando el rendimiento promedio y la desviación estándar
accuracy_mean = cross_val_scores.mean()
accuracy_std = cross_val_scores.std()

accuracy_mean, accuracy_std
```

```python
from sklearn.model_selection import GridSearchCV

# Definiendo el espacio de hiperparámetros para la Regresión Logística
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Valores de C (coeficiente de regularización)
    'penalty': ['l1', 'l2']        # Tipos de regularización
}

# Creando el modelo de Regresión Logística para usar con GridSearchCV
logistic_model_grid = LogisticRegression(max_iter=1000, random_state=42)

# Configuración de GridSearchCV
grid_search = GridSearchCV(estimator=logistic_model_grid, param_grid=param_grid, 
                           cv=10, scoring='accuracy', verbose=1, n_jobs=-1)

# Ejecución de GridSearchCV
grid_search.fit(X_train, y_train)

# Mejores hiperparámetros y rendimiento correspondiente
best_params = grid_search.best_params_
best_score = grid_search.best_score_

best_params, best_score
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score, roc_curve
import matplotlib.pyplot as plt

# Predicciones con el modelo optimizado en el conjunto de prueba
y_pred = grid_search.predict(X_test)
y_pred_proba = grid_search.predict_proba(X_test)[:, 1]

# Calculando métricas: AUC, Accuracy y Recall
auc_score = roc_auc_score(y_test, y_pred_proba)
accuracy = accuracy_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

# Creando la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Dibujando la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], color='darkorange', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

auc_score, accuracy, recall

```

```python
from sklearn.neural_network import MLPClassifier

# Configuración del modelo de Red Neuronal
# Utilizaremos una red con dos capas ocultas, cada una con 100 neuronas como punto de partida
neural_network_model = MLPClassifier(hidden_layer_sizes=(100, 100), 
                                     max_iter=500, 
                                     random_state=42)

# Entrenamiento del modelo de Red Neuronal
neural_network_model.fit(X_train, y_train)

# Evaluación del modelo en el conjunto de prueba
nn_accuracy = neural_network_model.score(X_test, y_test)

nn_accuracy
```

```python
from sklearn.metrics import roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def plot_roc_auc(model, X_test, y_test, model_name):
    # Predicciones de probabilidad
    probs = model.predict_proba(X_test)
    preds = probs[:, 1]

    # Calcular la curva ROC-AUC
    fpr, tpr, threshold = roc_curve(y_test, preds)
    roc_auc = auc(fpr, tpr)

    # Graficar
    plt.figure()
    plt.plot(fpr, tpr, 'b', label = f'AUC = {roc_auc:.2f}')
    plt.title(f'Curva ROC para {model_name}')
    plt.ylabel('Tasa Verdadero Positivo')
    plt.xlabel('Tasa Falso Positivo')
    plt.legend(loc = 'lower right')
    plt.show()
```

```python
def plot_confusion_matrix(model, X_test, y_test, model_name):
    # Predicciones
    predictions = model.predict(X_test)

    # Matriz de confusión
    cm = confusion_matrix(y_test, predictions, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)

    # Graficar
    disp.plot()
    plt.title(f'Matriz de Confusión para {model_name}')
    plt.show()
```

```python
# Para Random Forest
plot_roc_auc(best_model_rf, X_test, y_test, 'Random Forest')
plot_confusion_matrix(best_model_rf, X_test, y_test, 'Random Forest')

# Para XGBoost
plot_roc_auc(best_model_xgb, X_test, y_test, 'XGBoost')
plot_confusion_matrix(best_model_xgb, X_test, y_test, 'XGBoost')
```

```python
def calcular_rentabilidad(model, X_test, y_test):
    # Obtener las predicciones y la matriz de confusión
    predicciones = model.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_test, predicciones).ravel()

    # Calcular la rentabilidad
    rentabilidad = (tp + tn) * 100 - fp * 20 - fn * 500
    return rentabilidad

# Calcular la rentabilidad para Random Forest
rentabilidad_rf = calcular_rentabilidad(best_model_rf, X_test, y_test)
print(f"Rentabilidad de Random Forest: {rentabilidad_rf} €")

# Calcular la rentabilidad para XGBoost
rentabilidad_xgb = calcular_rentabilidad(best_model_xgb, X_test, y_test)
print(f"Rentabilidad de XGBoost: {rentabilidad_xgb} €")
```
