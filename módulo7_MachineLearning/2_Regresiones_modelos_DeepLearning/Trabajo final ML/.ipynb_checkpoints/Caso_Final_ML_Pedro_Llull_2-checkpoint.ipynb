```python
import pandas as pd

# Cargando el archivo de datos
file_path = 'online_shoppers_intention.csv'
data = pd.read_csv(file_path)

# Visualizando las primeras filas del conjunto de datos y la informaci√≥n general
data_head = data.head()
data_info = data.info()

data_head, data_info
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilo de los gr√°ficos
sns.set(style="whitegrid")

# Lista de variables para el an√°lisis visual
variables_to_plot = ['Administrative', 'Informational', 'ProductRelated', 
                     'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']

# Creando histogramas para las variables seleccionadas
plt.figure(figsize=(15, 10))
for i, var in enumerate(variables_to_plot):
    plt.subplot(3, 3, i+1)
    sns.histplot(data[var], kde=False, bins=30)
    plt.title(var)
    plt.tight_layout()

plt.show()
```

```python
# A√±adiendo variables de duraci√≥n a la lista de an√°lisis
variables_duration = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration']
variables_to_plot.extend(variables_duration)

# Creando histogramas para las variables de duraci√≥n
plt.figure(figsize=(18, 12))
for i, var in enumerate(variables_duration):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[var], kde=False, bins=30)
    plt.title(var)
    plt.tight_layout()

plt.show()
```

```python
# Creando boxplots para las variables seleccionadas
plt.figure(figsize=(18, 12))
for i, var in enumerate(variables_to_plot):
    plt.subplot(4, 4, i+1)
    sns.boxplot(x=data[var])
    plt.title(var)
    plt.tight_layout()

plt.show()
```

```python
# Eliminando filas con valores faltantes
data_cleaned = data.dropna()

# Verificando la eliminaci√≥n de valores faltantes
missing_values_after = data_cleaned.isnull().sum()

missing_values_after, data_cleaned.shape
```

```python
# Filtrando solo las columnas num√©ricas para el c√°lculo de IQR y valores at√≠picos
numeric_cols = data_cleaned.select_dtypes(include=['float64', 'int64'])

# C√°lculo del IQR para las variables num√©ricas
Q1_numeric = numeric_cols.quantile(0.25)
Q3_numeric = numeric_cols.quantile(0.75)
IQR_numeric = Q3_numeric - Q1_numeric

# Identificaci√≥n de valores at√≠picos en las columnas num√©ricas
outliers_numeric = ((numeric_cols < (Q1_numeric - 1.5 * IQR_numeric)) | (numeric_cols > (Q3_numeric + 1.5 * IQR_numeric))).sum()

outliers_numeric, IQR_numeric
```

```python
# Variables categ√≥ricas a convertir usando One-Hot Encoding
categorical_cols = data_cleaned.select_dtypes(include=['object', 'bool']).columns

# Aplicando One-Hot Encoding
data_dummies = pd.get_dummies(data_cleaned, columns=categorical_cols, drop_first=True)

# Mostrando las nuevas columnas creadas y el tama√±o del conjunto de datos modificado
data_dummies_head = data_dummies.head()
data_dummies_shape = data_dummies.shape

data_dummies_head, data_dummies_shape
```

```python
import numpy as np

# Calculando la matriz de correlaci√≥n
correlation_matrix = data_dummies.corr()

# Identificamos variables altamente correlacionadas (umbral de correlaci√≥n, por ejemplo, > 0.9)
high_correlation = np.where(np.abs(correlation_matrix) > 0.9)
high_correlation = [(correlation_matrix.columns[x], correlation_matrix.columns[y]) for x, y in zip(*high_correlation) if x != y and x < y]

high_correlation
```

```python
from sklearn.preprocessing import StandardScaler

# Creando el objeto StandardScaler
scaler = StandardScaler()

# Variables num√©ricas a estandarizar
numeric_cols = data_dummies.select_dtypes(include=['float64', 'int64']).columns

# Aplicando la estandarizaci√≥n a las variables num√©ricas
data_dummies[numeric_cols] = scaler.fit_transform(data_dummies[numeric_cols])

# Mostrando las primeras filas del conjunto de datos estandarizado
data_dummies.head()
```

```python
from sklearn.model_selection import train_test_split

# Definiendo las variables independientes (X) y la variable dependiente (y)
X = data_dummies.drop('Revenue_True', axis=1)
y = data_dummies['Revenue_True']

# Dividiendo los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Verificando las dimensiones de los conjuntos de datos resultantes
(X_train.shape, X_test.shape), (y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Configurando el modelo de Regresi√≥n Log√≠stica con un mayor n√∫mero de iteraciones
logistic_model_iter = LogisticRegression(max_iter=1000)  # Aumentando el n√∫mero de iteraciones

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],  # Valores de C para probar
    'penalty': ['l2'],  # Usando L2 como regularizaci√≥n
    'solver': ['lbfgs']  # Solver lbfgs es compatible con la regularizaci√≥n L2
}

# Creando el GridSearch con validaci√≥n cruzada
grid_search_iter = GridSearchCV(logistic_model_iter, param_grid, cv=5, scoring='accuracy')

# Ajustando el modelo
grid_search_iter.fit(X_train, y_train)

# Mejores par√°metros y mejor puntuaci√≥n con el aumento de iteraciones
best_params_iter = grid_search_iter.best_params_
best_score_iter = grid_search_iter.best_score_

best_params_iter, best_score_iter

```

```python
from sklearn.model_selection import cross_val_score

# Evaluando el modelo con la m√©trica ROC-AUC en validaci√≥n cruzada
roc_auc_scores_cv = cross_val_score(grid_search_iter.best_estimator_, X_train, y_train, cv=5, scoring='roc_auc')

# Promedio de las puntuaciones ROC-AUC en validaci√≥n cruzada
roc_auc_average_cv = roc_auc_scores_cv.mean()

roc_auc_average_cv
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score

# Predicciones en el conjunto de prueba
y_pred = grid_search_iter.predict(X_test)
y_pred_proba = grid_search_iter.predict_proba(X_test)[:, 1]

# Calculando m√©tricas en el conjunto de prueba
roc_auc_test = roc_auc_score(y_test, y_pred_proba)
accuracy_test = accuracy_score(y_test, y_pred)
recall_test = recall_score(y_test, y_pred)

roc_auc_test, accuracy_test, recall_test

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Creando el modelo de red neuronal
nn_model = Sequential()
nn_model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))  # Capa de entrada
nn_model.add(Dense(8, activation='relu'))  # Capa oculta
nn_model.add(Dense(1, activation='sigmoid'))  # Capa de salida

# Compilando el modelo
nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Resumen del modelo
nn_model.summary()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score

# Definici√≥n del modelo (suponiendo que ya tienes definido el modelo como 'nn_model')
# nn_model = Sequential()
# [Agrega aqu√≠ las capas del modelo, como lo has hecho antes]

# Compilaci√≥n del modelo
nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Entrenamiento del modelo
history = nn_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)

# Evaluaci√≥n del modelo en el conjunto de prueba
loss, accuracy = nn_model.evaluate(X_test, y_test, verbose=0)

# Predicciones en el conjunto de prueba
y_pred_proba = nn_model.predict(X_test).ravel()  # Probabilidades
y_pred = np.where(y_pred_proba > 0.5, 1, 0)      # Clases predichas

# C√°lculo de ROC-AUC y Recall
roc_auc = roc_auc_score(y_test, y_pred_proba)
recall = recall_score(y_test, y_pred)

# Imprimir las m√©tricas
print(f'Accuracy: {accuracy:.4f}')
print(f'ROC-AUC: {roc_auc:.4f}')
print(f'Recall: {recall:.4f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score

# Construyendo el modelo de Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Par√°metros para GridSearchCV
param_grid_rf = {
    'n_estimators': [100, 200, 300],  # N√∫mero de √°rboles
    'max_depth': [10, 20, 30],        # Profundidad m√°xima de los √°rboles
    'min_samples_split': [2, 5, 10]   # N√∫mero m√≠nimo de muestras requeridas para dividir un nodo
}

# Creando el GridSearchCV
grid_search_rf = GridSearchCV(rf_model, param_grid_rf, cv=5, scoring='accuracy', verbose=1)

# Ajustando el modelo
grid_search_rf.fit(X_train, y_train)

# Mejores par√°metros y mejor puntuaci√≥n
best_params_rf = grid_search_rf.best_params_
best_score_rf = grid_search_rf.best_score_

# Evaluando el modelo en el conjunto de prueba
y_pred_rf = grid_search_rf.predict(X_test)
y_pred_proba_rf = grid_search_rf.predict_proba(X_test)[:, 1]

# Calculando las m√©tricas
roc_auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)

# Imprimiendo los resultados
print("Mejores par√°metros:", best_params_rf)
print("Mejor puntuaci√≥n CV:", best_score_rf)
print("ROC-AUC en test:", roc_auc_rf)
print("Accuracy en test:", accuracy_rf)
print("Recall en test:", recall_rf)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generando las predicciones del modelo
y_pred_rf = grid_search_rf.predict(X_test)

# Calculando la matriz de confusi√≥n
conf_matrix = confusion_matrix(y_test, y_pred_rf)

# Visualizando la matriz de confusi√≥n
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues')
plt.xlabel('Predicciones')
plt.ylabel('Valores Verdaderos')
plt.title('Matriz de Confusi√≥n - Random Forest')
plt.show()
```

```python
# Definiendo los valores de la matriz de confusi√≥n
TP = conf_matrix[0][0]
TN = conf_matrix[1][1]
FP = conf_matrix[0][1]
FN = conf_matrix[1][0]

# Definiendo las ganancias y p√©rdidas
ganancia_TP = 100  # Ganancia por cada Verdadero Positivo
ganancia_TN = 100  # Ganancia por cada Verdadero Negativo
perdida_FP = 20    # P√©rdida por cada Falso Positivo
perdida_FN = 500   # P√©rdida por cada Falso Negativo

# Calculando la rentabilidad
rentabilidad = (TP * ganancia_TP) + (TN * ganancia_TN) - (FP * perdida_FP) - (FN * perdida_FN)
rentabilidad
```

```python
import xgboost as xgb

# Crear el modelo de XGBoost
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Entrenar el modelo
xgb_model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# Definir el espacio de b√∫squeda de hiperpar√°metros
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [4, 6, 8],
    'subsample': [0.8, 0.9, 1],
    'colsample_bytree': [0.8, 0.9, 1]
}

# Crear el objeto GridSearchCV
grid_search_xgb = GridSearchCV(xgb_model, param_grid_xgb, cv=5, scoring='accuracy')

# Ejecutar la b√∫squeda de hiperpar√°metros
grid_search_xgb.fit(X_train, y_train)

# Mejores par√°metros y mejor puntuaci√≥n
best_params_xgb = grid_search_xgb.best_params_
best_score_xgb = grid_search_xgb.best_score_

print("Mejores par√°metros:", best_params_xgb)
print("Mejor puntuaci√≥n:", best_score_xgb)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
# Evaluar el mejor modelo en el conjunto de prueba
best_model_xgb = grid_search_xgb.best_estimator_
predictions_xgb = best_model_xgb.predict(X_test)

# Imprimir el informe de clasificaci√≥n
print(classification_report(y_test, predictions_xgb))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Predicciones de probabilidad para XGBoost
probs_xgb = best_model_xgb.predict_proba(X_test)
preds_xgb = probs_xgb[:, 1]

# Calcular la curva ROC-AUC para XGBoost
fpr_xgb, tpr_xgb, threshold_xgb = roc_curve(y_test, preds_xgb)
roc_auc_xgb = auc(fpr_xgb, tpr_xgb)

# Graficar la curva ROC
plt.figure()
plt.plot(fpr_xgb, tpr_xgb, 'b', label = f'AUC = {roc_auc_xgb:.2f}')
plt.title('Curva ROC para XGBoost')
plt.ylabel('Tasa Verdadero Positivo')
plt.xlabel('Tasa Falso Positivo')
plt.legend(loc = 'lower right')
plt.show()
```

```python
def calcular_rentabilidad(model, X_test, y_test):
    # Obtener las predicciones y la matriz de confusi√≥n
    predicciones = model.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_test, predicciones).ravel()

    # Calcular la rentabilidad
    rentabilidad = (tp + tn) * 100 - fp * 20 - fn * 500
    return rentabilidad

# Calcular la rentabilidad para Random Forest
#rentabilidad_rf = calcular_rentabilidad(best_model_rf, X_test, y_test)
#print(f"Rentabilidad de Random Forest: {rentabilidad_rf} ‚Ç¨")

# Calcular la rentabilidad para XGBoost
rentabilidad_xgb = calcular_rentabilidad(best_model_xgb, X_test, y_test)
print(f"Rentabilidad de XGBoost: {rentabilidad_xgb} ‚Ç¨")
```
