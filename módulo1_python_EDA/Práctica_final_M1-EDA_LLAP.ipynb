__1-Une todos los dataframe desde Enero a Diciembre en un único dataframe. Muestra el shape total del dataframe resultante.__

```python
import pandas as pd
```

```python
#Ejemplo de leer un csv (después de varios errores se ve que hay un error de encoding utf-8, por lo que la cambié a 'latin1'
#y que tenga en cuenta el separador ';')

df = pd.read_csv("CSVs/202201_detalle.csv", sep = ";", encoding = "latin1")
df
```

```python
#Preparamos una lista vacía
dataframes = []

#Iterar sobre archivos CSV y cargar cada uno en un DataFrame separado utilizando la función read_csv
#de Pandas. Se puede utilizar la función glob de la biblioteca glob para obtener una lista de los
#archivos CSV en un directorio determinado:

import glob

#Ruta a los archivos CSV
ruta = "CSVs/*.csv"

#Itera sobre los archivos CSV, a tener en cuenta el separador y la codificación de caracteres "latinos"
for archivo_csv in glob.glob(ruta):
    df = pd.read_csv(archivo_csv, sep= ";", encoding= "latin1")
    dataframes.append(df)
```

```python
#Cargados todos los archivos CSV en DataFrames individuales y almacenados en la
#lista dataframes, los combinamos en un solo DataFrame utilizando la función concat de Pandas:

#Combina los DataFrames en uno solo
#El parámetro ignore_index=True se utiliza para restablecer los índices de fila en el DataFrame combinado.
df_combinado = pd.concat(dataframes, ignore_index=True)
```

__6-Genera una nueva columna llamada DIFERENCIA_KMH que sea el resultado de restar a cada valor de la velocidad a la que circulaba el conductor la velocidad límite en la que se produce la infracción, obtén un top 10 de los conductores que rebasaron los límites a mayor velocidad y un top 5 a menor velocidad.__

```python
#Calcular la columna "DIFERENCIA_KMH"
df_combinado3['DIFERENCIA_KMH'] = df_combinado3['VEL_CIRCULA '] - df_combinado3['VEL_LIMITE']

#Ordenar el DataFrame por la columna "DIFERENCIA_KMH" de manera descendente
df_combinado4 = df_combinado3.sort_values('DIFERENCIA_KMH', ascending=False)

#Obtener el top 10 de los conductores que rebasaron los límites a mayor velocidad
top_10_mv = df_combinado4.head(10)

#Mostrar los resultados
print("Top 10 de conductores que rebasaron los límites a mayor velocidad:")
print(top_10_mv)
```

```python
#Obtener el top 5 de los conductores que rebasaron los límites a menor velocidad
#Al obtener los resultados del anterior dataframe me salen valores a 0 y creo que no es la información
#que se esepera, ya que queremos saber los de infración a menor velocidad
top_5_menv = df_combinado4.tail(5)

print("Top 5 de conductores que rebasaron los límites a menor velocidad:") 
print(top_5_menv)
```

```python
#Para ello calcularemos de nuevo pero ignorando los resultados 0 y de forma ascendente
df_combinado3['DIFERENCIA_KMH'] = df_combinado3['VEL_CIRCULA '] - df_combinado3['VEL_LIMITE']

#Filtrar los registros con diferencias menores a 0 y distintas de 0
filtro = (df_combinado3['DIFERENCIA_KMH'] > 0) & (df_combinado3['DIFERENCIA_KMH'] != 0)
df_filtrado = df_combinado3.loc[filtro]

#Ordenar el DataFrame por la columna "DIFERENCIA_KMH" de manera ascendente
df_ordenado = df_filtrado.sort_values('DIFERENCIA_KMH')

#Obtener el top 5 de los conductores que rebasaron los límites a menor velocidad
top_5_menor_v = df_ordenado.head(5)

#Mostrar los resultados
print("Top 5 de conductores que rebasaron los límites a menor velocidad (distinta de 0):")
print(top_5_menor_v)
```

__7-Obtén un dataframe filtrado por todas las denuncias que hayan acarreado retirada de puntos. Agrupa el dataframe filtrado por el agente público que emite la denuncia, ¿cuál de todos los agentes públicos tiene una mayor media de retirada de puntos?__

```python
#Filtrar el DataFrame por denuncias que hayan acarreado retirada de puntos
df_filtrado = df_combinado3[df_combinado3[' PUNTOS'] > 0]

#Agrupar el DataFrame filtrado por el agente público y calcular la media de la retirada de puntos
agrupado = df_filtrado.groupby('DENUNCIANTE')[' PUNTOS'].mean()

#Encontrar el agente público con la mayor media de retirada de puntos
agente_mayor_media = agrupado.idxmax()

#Mostrar la información del agente público con la mayor media de retirada de puntos
print("Agente público con mayor media de retirada de puntos:", agente_mayor_media)
```

__8-De la columna hora elimina la parte decimal, muestra de forma gráfica en qué horas se han producido más multas.__

```python
import matplotlib.pyplot as plt

#Convertir la columna "Hora" a tipo entero
df_combinado3['HORA'] = df_combinado3['HORA'].astype(int)

#Contar la frecuencia de cada hora
frecuencia_horas = df_combinado3['HORA'].value_counts()

#Ordenar las horas de forma ascendente
frecuencia_horas = frecuencia_horas.sort_index()

#Crear un gráfico de barras para visualizar las horas con más multas
frecuencia_horas.plot(kind='bar')
plt.xlabel('HORA')
plt.ylabel('FRECUENCIA')
plt.title('MULTAS POR HORA')
plt.show()
```

__10-Muestra el número de infracciones de forma gráfica (sin frecuencia acumulada), de cada uno de los agentes públicos en función de la calificación de la infracción__

```python
#Agrupar los datos por 'DENUNCIANTE' y 'CALIFICACION' y contar el número de infracciones
infracciones_por_agente = df_combinado3.groupby(['DENUNCIANTE', 'CALIFICACION']).size().unstack()

#Crear un gráfico de barras para visualizar el número de infracciones por agente y calificación
infracciones_por_agente.plot(kind='bar', color= ['red', 'blue', 'purple'])
plt.xlabel('Agente Público')
plt.ylabel('Número de Infracciones')
plt.title('Número de Infracciones por Agente y Calificación')
plt.legend(title='Calificación de Infracción')
plt.show()
```
