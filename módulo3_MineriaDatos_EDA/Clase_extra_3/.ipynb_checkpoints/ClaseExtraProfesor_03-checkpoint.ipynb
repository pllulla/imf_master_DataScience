```python
paises = req.json()['countries']
```

```python
airports = []
url = "https://www.air-port-codes.com/api/v1/multi"
for pais in paises:
    
    try:
        data = {
        "term": pais['name']
        }

        req = requests.post(url, data=data, headers=headers)
        req = req.json()
    
        
        for airport in req['airports']:
            airports.append(airport)
    
    except Exception as e:
        print(e)
    
    
    
```

```python
len(airports)
```

```python
airports_clean = []

for airport in airports:
    
    airport['country_name'] = airport['country']['name']
    airport['country_code'] = airport['country']['iso']
    airport['state'] = airport['state']['name']
    del airport['country']
    
    airports_clean.append(airport)
    
```

```python
db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Pepito.08"
)
```

```python
cursor = db.cursor()
```

```python
cursor.execute("CREATE DATABASE aeropuertos")
cursor.execute("USE aeropuertos")
cursor.execute("""
    CREATE TABLE airports (
    name varchar(60),
    city varchar(60),
    iata char(3),
    latitude varchar(11),
    longitude varchar(11),
    state varchar(60),
    country_name varchar(60),
    country_code char(2)
    );
""")
```

```python
for airport in airports_clean:
    
    #Limpiamos los nombres de las columnas
    cols = ''
    
    #Recorremos las keys del objeto airport para obtener las columnas
    for col in airport.keys():
        #Para cada columna la guardamos en cols
        #concatenando los valores y separando por ,
        cols = cols + col + ', '
    
    # Borramos el espacio final y la última ,
    cols = cols.strip()[:-1]
        
    vals = ''
    
    # Se repite el proceso de las columnas con los valores
    for val in airport.values():
        vals = vals + "'" + val + "', "
    vals = vals.strip()[:-1]
        
    
    # Se insertan las columnas y los valores en la query SQL
    query = f"""
        INSERT INTO 
            airports ({cols})
        VALUES ({vals})
            """
    
    #Se ejecuta la query y se guarda la BDD
    cursor.execute(query)
    db.commit()
```
