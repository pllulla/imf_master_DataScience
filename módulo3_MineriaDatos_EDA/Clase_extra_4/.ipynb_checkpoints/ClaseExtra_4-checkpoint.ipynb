__Para este ejercicio necesitarás crear tres dataframes que contenga datos de la geografía española__

    Los dataframes son los siguientes y deben contener la siguiente información:

    Estado (comunidades autónomas)
        id -> Número secuencial que comienza por 1001
        countryCode -> En nuestro caso siempre "ES"
        nombre -> Nombre del estado

    región (provincia)
        id -> Número secuencial que comienza por 1001
        stateID -> Id del estado propietario
        nombre -> Nombre de la región

    Ciudades
        id -> Número secuencial que comienza por 1001
        regionID -> Id del propietario Región
        Nombre -> Nombre de la ciudad
        Habitantes-> Rango de residentes en la ciudad

    Notas:
    Los datos deben ser extraídos de wikipedia realizando scrapping: https://es.wikipedia.org/wiki/Anexo:Municipios_de_Espa%C3%B1a_por_poblaci%C3%B3n

```python
import mysql.connector
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
webpage = 'https://es.wikipedia.org/wiki/Anexo:Municipios_de_Espa%C3%B1a_por_poblaci%C3%B3n'

req = Request(webpage, headers={'User-Agent': 'Mozilla/5.0'})
raw_web = urlopen(req, timeout=10).read()
```

```python
soup = BeautifulSoup(raw_web, 'html.parser')
tables = soup.find_all('table')
titulos_tablas = soup.find_all('span', attrs={"class": "mw-headline"})
```

```python
print('Títulos de tablas: ', len(tables))
print('Tablas: ', len(tables))
```

```python
head = []
rows_raw = []
for table in tables:
    
    head = table.find_all('th') #We skip the first header as we know we don't want it
    table_rows = table.find_all('tr')[1:]
    
    for fila in table_rows:
        cols = []
        for col in fila.find_all('td'):
            cols.append(col.text.strip())
            
        rows_raw.append(cols)
```

```python
rows_raw
```

```python
cleaned_data = []

for row in rows_raw:
    cleaned_row = []
    
    # Eliminar espacios en blanco adicionales y convertir a entero en la primera columna
    cleaned_row.append(int(row[0])+1000)
    
    # Eliminar espacios y caracteres no deseados en los campos de texto
    cleaned_row.append(row[1].strip())
    cleaned_row.append(int("".join(row[2].split())))
    cleaned_row.append(row[3].split("\xa0")[0])
    
    if len(row) > 4:
        cleaned_row.append(row[4].split("\xa0")[0])
    
    cleaned_data.append(cleaned_row)

# Mostrar los datos limpios
for row in cleaned_data:
    print(row)
```

```python
all_data = pd.DataFrame(cleaned_data, columns=['ID', 'City', 'Population', 'Region', 'State'])
all_data
```

Estado (comunidades autónomas)
* id -> Número secuencial que comienza por 1001
* countryCode -> En nuestro caso siempre "ES"
* nombre -> Nombre del estado

```python
# Generar la tabla State

id_c = all_data['ID']
states_c = all_data['State'].drop_duplicates()

states_list = []
for idd, statee in zip(id_c, states_c):
    
    states_list.append({"id": idd, "name": statee, "country": "ES"})

states = pd.DataFrame(states_list, columns = states_list[0].keys())

states.head()
```

región (provincia)
* id -> Número secuencial que comienza por 1001
* stateID -> Id del estado propietario
* nombre -> Nombre de la región

```python
# Generar la tabla region

id_c = all_data['ID']
regions_c = all_data['Region'].drop_duplicates()

regions_list = []
for iddd, region in zip(id_c, regions_c):
    
    regions_list.append({"id":iddd, "stateID": iddd, "name": region})
    
regions = pd.DataFrame(regions_list, columns=regions_list[0].keys())

regions.head()
```

Ciudades
* id -> Número secuencial que comienza por 1001
* regionID -> Id del propietario Región
* Nombre -> Nombre de la ciudad
* Habitantes-> Rango de residentes en la ciudad

```python
# Generar la tabla poblaciones

id_c = all_data['ID']
city_c = all_data['City'].drop_duplicates()
popu_c = all_data['Population']

cities_list = []
for idddd, ciudad, population in zip(id_c, city_c, popu_c):
    
    cities_list.append({"id": idddd, "regionID": idddd, "name": ciudad, "population": population})
    
cities = pd.DataFrame(cities_list, columns=cities_list[0].keys())

cities.head()
```
