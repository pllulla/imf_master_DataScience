```python
#Web scrapping: Hacerse pasar por un usuario a una web y extraer la información de esa web (lo que está a la vista)
```

```python
#Necesitamos REQUEST, URLOPEN y BEAUTIFULSOUP
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Metemos la url que nos interesa en una variable
url = "https://es.wikipedia.org/wiki/Anne_Hathaway"
#Engañamos la web como usuarios
headers = {"User-Agent":"Mozilla/5.0"}
```

```python
req = Request(url, headers=headers)
html = urlopen(req).read()
html
```

```python
#Con BEAUTIFULSOUP tenemos que desgranar el contenido y extraer la información
soup = BeautifulSoup(html, 'html.parser')
```

```python
#Vamos a ver los titulos de la web
soup.find_all('h3')
```

```python
#Vamos a ver la tabla que nos interesa, cuántas?
len(soup.find_all('table'))
```

```python
#Solo nos interesan 2 tablas de filmografias
#Hemos buscado por la clase de la tabla en el inspector de la página web y ese nos da fallo:
#class="wikitable sortable jquery-tablesorter"
#No nos sale pq el código que muestra la página no tiene pq coincidir en cada navegador y en python
#Mejor buscar en el primer código descagado en la variable html, buscar el CLASS alli para asegurar
#Vemos en el código en bruto en la variable HTML que hemos descargado al principio que la clase esta definida
#de forma diferente
tables = soup.find_all('table', attrs={'class':'wikitable sortable'})
len(tables)
```

```python
tables
```

```python
#Queremos 1 tabla, la primera pq hay 2
tabla_html_pelis = tables[0]
```

```python
tabla_html_pelis
```

```python
#TR son filas y TD son columnas
len(tabla_html_pelis.find_all('tr'))
```

```python
filas = tabla_html_pelis.find_all('tr')
filas[0]
```

```python
filas = tabla_html_pelis.find_all('tr')
filas[1]
```

```python
filas[1].find_all('th')
```

```python
#Filtramos hasta limpiar los datos al detalle

columnas = []
for columna in filas[1].find_all('th'):
    columnas.append(columna.text.strip())
    
columnas
```

```python
#Recoremos las filas, que ya teniamos antes
filas[2:]
```

```python
#
filas_clean = []
for fila in filas[2:]:
    
    row = []
    for col in fila.find_all('td'):
        row.append(col.text.strip())
    filas_clean.append(row)
    
filas_clean
```

```python
#Para cuadrar tenemos que meter el año en cada columna, pq hay años que hizo varias películas y solo
#se expresa en la primera pelicula del año
filas_fix = []
for i in range(0, len(filas_clean)):

    fila = filas_clean[i]
    
    if len(fila) == 4:
        filas_fix.append(fila)
    elif len(fila) == 3:
        fix = [
            filas_fix[-1][0]
        ]
            
        for col in fila:
            fix.append(col)
            
        filas_fix.append(fix)
        
filas_fix
```

```python
pd.DataFrame(filas_fix, columns=columnas)
```
